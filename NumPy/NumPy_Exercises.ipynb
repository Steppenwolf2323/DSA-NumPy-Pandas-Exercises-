### 2. Transform a list, a tuple, and a set into NumPy arrays.

```python
ex_list = [1,2,3]
ex_tuple = ("a","b","c")
ex_set = {1.5, 4.33, 6.292}

print(type(ex_list),type(ex_tuple),type(ex_set), sep='  ')
print(ex_list,ex_tuple,ex_set, sep='  ')
print('\n')

new_list = np.array(ex_list)
new_tuple = np.array(ex_tuple)
new_set = np.array(ex_set)

print(type(new_list),type(new_tuple),type(new_set), sep='  ')
print(new_list,new_tuple,new_set)



```

### 3. Create a 0D array (scalar), a 1D array (vector) a 2D array (matrix) a 3D array (tensor).

```python
zero_dim = np.array(13)
one_dim = np.array([12,88])
two_dim = np.array([[12,44,16],[25,88,32]])
three_dim = np.array([[[10,40],[33,63]],[[40,10],[63,33]]])

# check on the dimension of the arrays usign '.ndim'

print(zero_dim.ndim, one_dim.ndim, two_dim.ndim, three_dim.ndim)
```

#### 4. Convert the 2D array above to a 3D array. Use the reshape method.

```python
new_two_dim = two_dim.reshape(2,1,3)
print(two_dim)
print("\n")
print(new_two_dim)
```

#### 5. Find the memory size of the 2D array above. Use the nbytes attribute.

```python
print(two_dim.nbytes)

print(new_two_dim.nbytes)
```

#### 6. Create a 4X4 array of random integers. Print the first 2 rows and the last 2 columns.

```python
matrix = np.random.randint(0,20, size=(4,4))

print(matrix)


print(f"First two rows:\n", matrix[0:2])

print(f"Last two columns:\n",matrix[:,-2:])

```

#### 7. Create a 5X5 array of random floats. Extract an array of elements from odd rows and even columns from it.

```python
big_m = np.random.randint(0,20, size=(5,5))
print(big_m, f'\n')

small_m = big_m[1::2,0::2]
print(small_m)
```

#### 8. Stack two arrays vertically and horizontally. Use the vstack and hstack functions.

```python
first_vec = np.random.randint(0,10, size=(3))
sec_vec = np.random.randint(0,10, size=(3))

print(first_vec, sec_vec, f'\n')

vertical = np.vstack((first_vec,sec_vec))

print(vertical, f"\nDimension: {vertical.ndim}\n")

horizontal = np.hstack((first_vec,sec_vec))
print(horizontal,f"\nDimension: {horizontal.ndim}")
```
